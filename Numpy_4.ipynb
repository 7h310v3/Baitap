Câu 1: Sử dụng thư viện Numpy, hãy tạo một vector a có 7 chiều (tương ứng 7 phần tử) với giá
trị các phần tử được sinh bằng hàm (kết quả xuất ra màn hình):
- np.arrange()
- np.linspace()
- np.logspace()
- np.random.rand()
Ghi chú: sau mỗi lần sinh ra giá trị cho vector, hãy cho biết số chiều của vector được sinh ra và
số chiều của vector chuyển vị của vector đó; cho biết kích thước của vector a và vector chuyển vị
của vector a.
Gợi ý: tham khảo ndim và shape.

```python
import numpy as np

# Using arange() function
A = np.arange(1, 8, 1)
print("Arange function: ",A)

# Using linespace() function
B = np.linspace(1, 7, 7, dtype = int)
print("Linespace function: ", B)

# Using logspace() function
C = np.logspace(1, 2, 7, dtype = int)
print("Logspace function: ", C)

# Using random.rand() function
D = np.random.rand(1, 7)
print("Rand function: ", D)
```

Câu 2: Sử dụng thư viện Numpy, hãy tạo ra ma trận A10x7 với giá trị các phần tử của ma trận A
được sinh bởi:
- np.arrange()
- np.linspace()
- np.logspace()
- np.random.rand()
Sau mỗi lần ma trận được tạo ra, hãy xuất kết quả ra màn hình và cho biết:
- Số chiều và kích thước của ma trận A và ma trận AT

```python
A = np.arange(0,70).reshape(10,7)
print(A)
print("Số chiều của vectoer a: ",A.ndim)
print("Số chiều vector chuyển vị vector a: ",A.T.ndim)
print("Kích thước vector A: ",A.shape)
print("Kích thước vector chuyển vị vector A: ",A.T.shape)

B = np.linspace(0,7,70).reshape(10,7)
print (B)
print("Số chiều của vector B: ",B.ndim)
print("Số chiều vector chuyển vị vector B: ",B.T.ndim)
print("Kích thước vector B: ",B.shape)
print("Kích thước vector chuyển vị vector B: ",B.T.shape)

C = np.logspace(1, 2, num = 70).reshape(10,7)
print (C)
print("Số chiều của vector a: ",C.ndim)
print("Số chiều vector chuyển vị vector C: ",C.T.ndim)
print("Kích thước vector C: ",C.shape)
print("Kích thước vector chuyển vị vector C: ",C.T.shape)

D = np.random.rand(10,7)
print(D)
print("Số chiều của vector D: ",D.ndim)
print("Số chiều vector chuyển vị vector D: ",D.T.ndim)
print("Kích thước vector D: ",D.shape)
print("Kích thước vector chuyển vị vector D: ",D.T.shape)
```

Câu 5: Thực hiện các yêu cầu sau:
- Xây dựng hàm tính giá trị trung bình
- Xây dựng hàm tính giá trị trung vị
- Xây dựng hàm tính giá trị yếu vị
- Xây dựng hàm tính phương sai
- Xây dựng hàm tính độ lệch chuẩn
- Hãy tạo ra ma trận A15x10 với các giá trị sinh bởi hàm arrange()
- Áp dụng 5 hàm ở trên cho từng hàng của ma trận A
- Áp dụng 5 hàm ở trên cho từng cột của ma trận A
Gợi ý: tham khảo hàm apply_along_axis() đã học ở bài trước.

```python
#5
A = np.arange(1,151).reshape(15,10)
print(A, "\n")

print("Giá trị trung bình từng hàng của A = \n", np.apply_along_axis(np.mean, 1, A))
print("Giá trị trung bình từng cột của A = \n", np.apply_along_axis(np.mean, 0, A))

print("Giá trị trung vị từng hàng của A = \n", np.apply_along_axis(np.median, 1, A))
print("Giá trị trung vị từng cột của A = \n", np.apply_along_axis(np.median, 0, A))

print("Phương sai từng hàng của A = \n", np.apply_along_axis(np.var, 1, A))
print("Phương sai từng cột của A = \n", np.apply_along_axis(np.var, 0, A))

print("Độ lệch chuẩn từng hàng của A = \n", np.apply_along_axis(np.std, 1, A))
print("Độ lệch chuẩn từng cột của A = \n", np.apply_along_axis(np.std, 0, A))

```
