Câu 1: Tạo các Series với kích thước 1x10 thỏa mãn yêu cầu:
- Series 1: có tên là “Họ và tên” và dữ liệu (định tính) do bạn tự nhập vào
- Series 2: có tên là “Ngày sinh” & dữ liệu do bạn tự nhập vào
- Series 3: có tên là “Tiền công/h” có dữ liệu định lượng do bạn tự nhập
- Series 4: có tên là “Số h công” có dữ liệu định lượng do bạn tự nhập
- Series 5: có tên là “Thực lĩnh” có kích thước 1x10 và để trống

Hãy:
- Tạo 1 dataframe bằng cách ghép các series trên lại với nhau
- Hãy tính giá trị cho cột (column) “Thực lĩnh” theo công thức:
Thực lĩnh = “Số h công” x “Tiền công/h”

- Hãy in toàn bộ dataframe ở trên ra màn hình

```python
import pandas as pd
import string
import random

ho = ["Đặng", "Lê", "Thái", "Nguyễn"]
dem = ["Nhật", "Thị", "Trọng", "Tâm"]
ten = ["Minh ", "Quân", "Xuân", "Thanh"]
ns = ["22/02/2002", "19/02/2002", "26/03/2002", "09/09/2002", "08/02/2002", "08/02/2003", "30/09/2002", "23/7/2002", "24/09/2003","09/06/2003"]

hvt = [random.choice(ho)+" "+random.choice(dem)+" "+random.choice(ten) for i in range(10)]
tch = [random.randint(10000, 20000) for i in range(10)]
h = [random.randint(1, 10) for i in range(10)]
```

```python
sr1 = pd.Series(hvt)
sr2 = pd.Series(ns)
sr3 = pd.Series(tch)
sr4 = pd.Series(h)
thuc_linh = [sr3[i]*sr4[i] for i in range(10)]
sr5 = pd.Series(thuc_linh)

data = {
    "Họ và tên":sr1,
    "Ngày sinh":sr2,
    "Tiền công 1 giờ": sr3,
    "Số giờ công": sr4,
    "Thực lĩnh": sr5
}
df = pd.DataFrame(data)
df
```


Câu 2: Khởi tạo dataframe bằng các phương pháp khác nhau, hãy:
- Tạo dataframe có kích thước 10x3 từ cấu trúc list tương ứng, với dữ liệu cho dataframe
này gồm các cột Họ tên, Tuổi và Giới tính. Xuất kết quả ra màn hình;


```python
tuoi = [random.randint(18, 30) for i in range(10)]
gt = [random.choice(["Nam", "Nữ"]) for i in range(10)]

data = {
    "Họ tên": hvt,
    "Tuổi": tuoi,
    "Giới tính": gt
}

df = pd.DataFrame(data)
df
```

- Tạo dataframe như trên nhưng từ cấu trúc Dictionary có chứa các List giá trị tương ứng với key. Xuất kết quả ra màn hình

```python
hoa = ["hồng", "mai", "cúc", "vạn thọ", "sen", "bông giấy"]
qua = ["mận", "lê", "đào", "táo", "sầu riêng", "dừa"]

data = {
    "Hoa": hoa,
    "Quả": qua
}

df = pd.DataFrame(data)
df
```

- Tạo dataframe như trên nhưng từ cấu trúc List của các dictionary. Xuất kết quả ra màn hình;

```python
df = pd.DataFrame(data.values())
df
```

- Tạo 3 list (1x10) tương ứng với Họ tên, Tuổi và Giới tính. Sử dụng hàm zip() để tạo dataframe từ 3 list riêng biệt này. Xuất kết quả ra màn hình.

```python
df = pd.DataFrame(zip(hvt, tuoi, gt))
df
```
